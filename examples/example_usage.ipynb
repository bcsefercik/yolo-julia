## Example Usage

### Activate Environment

```julia
import Pkg

Pkg.activate("../Project.toml")

img_size = (416, 416)
```

### Include libs

```julia
include("../models.jl")
include("../utils/load_darknet_weights.jl")
include("../utils/coco2014_utils.jl")
include("../utils/img.jl")
```

### Create a Darknet Instance and Load Weights

```julia
darknet = Darknet("../cfg/yolov3.cfg"; img_size=img_size);
load_darknet_weights(darknet, "../yolov3.weights")
```

### Load an image, predict and visualize

```julia
dog = read_image("../data/dog.jpg");
dog = reshape(dog, (size(dog)..., 1));

# Prediction
dog_out = darknet(dog, training=false);

#Visualization
img = draw_results(dog[:, :, :, 1], dog_out[1])
```
